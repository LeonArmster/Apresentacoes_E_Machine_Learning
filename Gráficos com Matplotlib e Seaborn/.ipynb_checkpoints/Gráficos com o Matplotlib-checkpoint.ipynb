### Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x**2
```

```python
# GrÃ¡fico com a cor Red
plt.plot(x, y, 'red')

# Nome do eixo X
plt.xlabel('Eixo X')

# Nome do eixo y
plt.ylabel('Eixo Y')

# TÃ­tulo do GrÃ¡fico
plt.title('TÃ­tulo')
```

```python
# FunÃ§Ã£o subplot para criaÃ§Ã£o de 2 eixos num mesmo grÃ¡fico

plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

### OrientaÃ§Ã£o Ã  objeto no Matplotlib

```python
# Exemplo de criaÃ§Ã£o mais manual
# Vantagens: melhores definiÃ§Ãµes para criaÃ§Ã£o de grÃ¡ficos

fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #primeiro % de posicionamento do grÃ¡fico, segundo = parte inferior terceiro = largura, quarto valor = altura

axes.plot(x, y, 'blue')

axes.set_xlabel('label x') # eixo X
axes.set_ylabel('label y') # eixo y
axes.set_title('titulo') # Titulo
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes.plot(x, y, 'blue')

axes.set_xlabel('label x') # eixo X
axes.set_ylabel('label y') # eixo y
axes.set_title('titulo') # Titulo

axes2.plot(y, x, 'red')

axes2.set_xlabel('label x2') # eixo X
axes2.set_ylabel('label y2') # eixo y
axes2.set_title('titulo2') # Titulo
```

### Subplots

```python
#Subplots atuam com desempacotamento de tuplas
#fig = elementos da figure
#ax = elementos dos eixos

fig, ax = plt.subplots()

ax.plot(x, y, 'r--')
ax.set_title('Titulo')
```

```python
# Passando parÃ¢metros para o subplot

fig, ax = plt.subplots(nrows = 1, ncols = 2)

for axis in ax:
    axis.plot(x, y, 'r')
```

```python
# Passando parÃ¢metros para o subplot

fig, ax = plt.subplots(nrows = 1, ncols = 2)

ax[0].plot(x,y, 'red')
ax[1].plot(x,y, 'g--')
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)

for axis in ax:
    axis.plot(x, y, 'r')
    
plt.tight_layout() ## ajusta na tela
```

```python
# funÃ§Ã£o para configurar o tamanho do grÃ¡fico

fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (30,8))

for axis in ax:
    axis.plot(x, y, 'r')
    
```

### Legendas, RÃ³tulos e TÃ­tulos

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, 'b--', label = 'x^2', alpha = 0.5)
ax.plot(x, x**3, 'g*-', label = 'x^3')

ax.set_title('TÃ­tulo') #TÃ­tulo
ax.set_xlabel('Eixo X') #Nome do eixo X
ax.set_ylabel('Eixo Y') #Nome do eixo Y
ax.legend(loc = 6) #Legenda
```

```python
fig, axes = plt.subplots(1, 3, figsize = (12,4))

axes[0].plot(x, x**2)
axes[0].plot(x, x**3)
axes[0].set_title('Intervalo de eixos padrÃ£o')

axes[1].plot(x, x**2)
axes[1].plot(x, x**3)
axes[1].axis('tight') ## Deixando a exibiÃ§Ã£o mais adequado e "apertado" 
axes[1].set_title('Eixos Apertados')

axes[2].plot(x, x**2)
axes[2].plot(x, x**3)
axes[2].set_ylim([0, 60]) ## Determinando qual parte serÃ¡ visualizada do grÃ¡fico, ou seja, onde comeÃ§ara e terminarÃ¡ o eixo Y
axes[2].set_xlim([2, 5]) ## Determinando qual parte serÃ¡ visualizada do grÃ¡fico, ou seja, onde comeÃ§ara e terminarÃ¡ o eixo X
axes[2].set_title('Intervalo de eixos padrÃ£o')
```
