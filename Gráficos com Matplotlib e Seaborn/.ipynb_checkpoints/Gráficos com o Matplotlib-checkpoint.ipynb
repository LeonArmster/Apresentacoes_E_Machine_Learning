### Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x**2
```

```python
# Gráfico com a cor Red
plt.plot(x, y, 'red')

# Nome do eixo X
plt.xlabel('Eixo X')

# Nome do eixo y
plt.ylabel('Eixo Y')

# Título do Gráfico
plt.title('Título')
```

```python
# Função subplot para criação de 2 eixos num mesmo gráfico

plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

### Orientação à objeto no Matplotlib

```python
# Exemplo de criação mais manual
# Vantagens: melhores definições para criação de gráficos

fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #primeiro % de posicionamento do gráfico, segundo = parte inferior terceiro = largura, quarto valor = altura

axes.plot(x, y, 'blue')

axes.set_xlabel('label x') # eixo X
axes.set_ylabel('label y') # eixo y
axes.set_title('titulo') # Titulo
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes.plot(x, y, 'blue')

axes.set_xlabel('label x') # eixo X
axes.set_ylabel('label y') # eixo y
axes.set_title('titulo') # Titulo

axes2.plot(y, x, 'red')

axes2.set_xlabel('label x2') # eixo X
axes2.set_ylabel('label y2') # eixo y
axes2.set_title('titulo2') # Titulo
```

### Subplots

```python
#Subplots atuam com desempacotamento de tuplas
#fig = elementos da figure
#ax = elementos dos eixos

fig, ax = plt.subplots()

ax.plot(x, y, 'r--')
ax.set_title('Titulo')
```

```python
# Passando parâmetros para o subplot

fig, ax = plt.subplots(nrows = 1, ncols = 2)

for axis in ax:
    axis.plot(x, y, 'r')
```

```python
# Passando parâmetros para o subplot

fig, ax = plt.subplots(nrows = 1, ncols = 2)

ax[0].plot(x,y, 'red')
ax[1].plot(x,y, 'g--')
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)

for axis in ax:
    axis.plot(x, y, 'r')
    
plt.tight_layout() ## ajusta na tela
```

```python
# função para configurar o tamanho do gráfico

fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (30,8))

for axis in ax:
    axis.plot(x, y, 'r')
    
```

### Legendas, Rótulos e Títulos

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, 'b--', label = 'x^2', alpha = 0.5)
ax.plot(x, x**3, 'g*-', label = 'x^3')

ax.set_title('Título') #Título
ax.set_xlabel('Eixo X') #Nome do eixo X
ax.set_ylabel('Eixo Y') #Nome do eixo Y
ax.legend(loc = 6) #Legenda
```

```python
fig, axes = plt.subplots(1, 3, figsize = (12,4))

axes[0].plot(x, x**2)
axes[0].plot(x, x**3)
axes[0].set_title('Intervalo de eixos padrão')

axes[1].plot(x, x**2)
axes[1].plot(x, x**3)
axes[1].axis('tight') ## Deixando a exibição mais adequado e "apertado" 
axes[1].set_title('Eixos Apertados')

axes[2].plot(x, x**2)
axes[2].plot(x, x**3)
axes[2].set_ylim([0, 60]) ## Determinando qual parte será visualizada do gráfico, ou seja, onde começara e terminará o eixo Y
axes[2].set_xlim([2, 5]) ## Determinando qual parte será visualizada do gráfico, ou seja, onde começara e terminará o eixo X
axes[2].set_title('Intervalo de eixos padrão')
```

```python
## Gerando números aleatórios

from random import sample

data = sample(range(1, 1000), 100)
```

```python
## Gráfico de Histograma

plt.hist(data)
```

```python
## Gerando dados aleatórios para o boxplot

data = [np.random.normal(0, std, 100) for std in range(1,4)]
```

```python
## Gráfico de BoxPlot

plt.boxplot(data)
```
