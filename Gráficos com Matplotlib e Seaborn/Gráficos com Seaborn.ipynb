## Gráficos com Seaborn

O seaborn tem um foco muito mais estátistico e utiliza a matplotlib como plano de fundo para evoluir e trazer melhores visualizações gráficas

### Importando Bibliotecas

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.distplot(tips['total_bill'])
```

```python
# KDE = linha de análise
# bins = parâmetro de distribuição

sns.distplot(tips['total_bill'], kde = False, bins = 30)
```

```python
# analisar 2 variáveis ao mesmo tempo criando 2 distplots e +1 gráfico explicando o paramêtro do meio

sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# analisar 2 variáveis ao mesmo tempo criando 2 distplots e +1 gráfico explicando o paramêtro do meio
# Por padrão ele usará o tipo scatter, mas ao utilizar "kind" você pode definir o tipo de gráfico de cruzamento
# se kind = reg então ele fará uma regressão linear
# o gráfico mostrará uma correlação positiva

sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg')
```

```python
# analisar 2 variáveis ao mesmo tempo criando 2 distplots e +1 gráfico explicando o paramêtro do meio
# Por padrão ele usará o tipo scatter, mas ao utilizar "kind" você pode definir o tipo de gráfico de cruzamento
# se kind = hex então ele fará uma análise de "calor"
# o ponto mais escuro mostrará onde tem mais dados concentrados

sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
```

```python
# o Pairplot exibirá basicamente todos os cruzamentos numéricos do dataframe
# hue faz a segregação de dados através de uma coluna classificadora
# palette é a paleta de cores


sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

```python
## Rugplot exibi apenas traços em todas as aparições dos pontos marcados

sns.rugplot(tips['total_bill'])
```

```python
## KDE faz estimativas de através de uma distribuição "normal" inventada

sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

### Plots Categóricos

```python
## Barplot é o gráfico de barras

sns.barplot(x = 'sex', y = 'total_bill', data = tips)
```

```python
## Barplot é o gráfico de barras
## Estimator vai me mostrar como quero a distribuição dos dados através de funções que posso lançar

sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.std)
```

```python
## Countplot vai me mostrar um gráfico de barras para contagem

sns.countplot(x = 'sex', data = tips)
```

```python
## Boxplot é para exibição de outliers

sns.boxplot(x = 'day', y = 'total_bill', data = tips, palette = 'rainbow')
```

```python
## Boxplot é para exibição de outliers
## orient serve para ver os gráficos de maneria orizontal, porém para isso teria que ser feito a inversão das variáveis o y vira x e o x vira y 

sns.boxplot(y = 'day', x = 'total_bill', data = tips, palette = 'rainbow', orient = 'h')
```

```python
# Boxplot é para exibição de outliers
# hue faz a segregação de dados através de uma coluna classificadora

sns.boxplot(x = 'day', y = 'total_bill', data = tips, palette = 'rainbow', hue = 'sex')
```

```python
# Violinplot é igual ao boxplot com a diferença de exibição de KDEs
# Split ele dividirá o violino em dois enconimzando informação

sns.violinplot(x = 'day', y = 'total_bill', data = tips, palette = 'rainbow', hue = 'sex', splitt = True)
```

```python
# Swarmplot exibe a distribuição de dados por colunas por pontos

sns.swarmplot(x = 'day', y = 'total_bill', data = tips)
```

### Plots de Regressão

```python
# Gráfico de regressão linear
# hue faz a segregação de dados através de uma coluna classificadora
# markers são os tipos pelos quais os dados serão serão separados
# scatter_kws são as configurações de tamanho

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o', 'v'], scatter_kws = {'s':100})
```

```python
# Gráfico de regressão linear
# hue faz a segregação de dados através de uma coluna classificadora
# col criará uma quebra em colunas fazendo a separação pelo classificador informado
# row criará uma quebra em linhas fazendo a separação pelo classificador informado

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', col = 'time', row = 'day')
```

```python
# Gráfico de regressão linear
# hue faz a segregação de dados através de uma coluna classificadora
# markers são os tipos pelos quais os dados serão serão separados
# scatter_kws são as configurações de tamanho
# aspect informará o tamanho lateral do gráfico
# size informará o tamanho da altura do gráfico

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o', 'v'], scatter_kws = {'s':100}, aspect = 1.6, size = 8)
```

### Plots Matriciais (Heatmaps)

É utilizado quando eu tenho duas variáveis categóricas e queremos ver uma terceiza variável continua para representar no eixo Z

```python
voos = sns.load_dataset('flights')
```

```python
voos
```

```python
din_voo = voos.pivot_table(index = 'year', columns = 'month', values = 'passengers')
```

```python
# Mapa de calor(heatmap)
# É possível determinar o tamanho do gráfico fazendo a união com as funções do matplotlib
# annot exibe os valores das zonas de calor
# fmt = configuração da quantidade de casas decimais
# linewidths = linhas separadoras do gráfico
# cmap = Padrão de cores do heatmap

fig, ax = plt.subplots(figsize=(20, 8))
sns.heatmap(din_voo, ax = ax, annot = True, fmt = '.0f', linewidths = 1, cmap = 'magma')
```

```python
# O mapa de cluster reorganiza o dataframe para deixar os valores mais relacionados dentro das caixinhas e outros menos relacionados afastados

sns.clustermap(din_voo)
```
